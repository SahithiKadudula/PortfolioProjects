```python
#Import Libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline                            
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create

#Matplotlib inline is a feature that allows you to display Matplotlib graphs and charts directly in the Jupyter Notebook interface

#Read in the data

df = pd.read_csv(r"C:\Users\sripr\Documents\Downloads\Portfolio_Project\movies.csv")
```

```python
#Look at the data

df.head(10)
```

```python
#looking at missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
# removing null values
df.dropna(inplace=True)

# Print the DataFrame info to verify that null values are removed
print(df.info())
```

```python
#data types for the columns
df.dtypes
```

```python
#changing datatypes of the columns 
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

#checking the dataframe whether the datatype updated or not
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
#just for reference
#checking for the duplicates for specific column
#df['company'].sort_values(ascending=False)

#dropping the duplicates for specific column
#df['company'].drop_duplicates().sort_values(ascending=False)


#dropping duplicates in the entire dataframe
df.drop_duplicates()
df.head()
```

```python
# Budget - High Correaltion
# Company - High Cprrelation

#Scatterplot with Budget Vs Gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget Vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Movie')
plt.show()
```

```python
#Plot the Budget Vs Gross using seaborn
sns.regplot(x='budget', y='gross', data = df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
#Looking at correlation - works only for numeric fields

numeric_df = df.select_dtypes(include=['number'])
numeric_df.corr()
```

```python
#Pearson, Kendall, Spearman
numeric_df.corr(method='pearson')
```

```python
#High Correlation between Budget and Gross
```
